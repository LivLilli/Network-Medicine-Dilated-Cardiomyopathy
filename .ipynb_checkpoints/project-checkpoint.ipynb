### Clean Summary Dataframe

The TMPO gene has an irregular UniProt value. Infact it has two entries separated by ';'. For this reason we change this row, mantaining just one of the two entry. Thenwe add a new row for the same gene but with the other value of entry. At the end we will have for the same gene, 2 rows, which differ only by the UniProt value.

```python
# looking at TMPO gene row
summary_df[summary_df['Gene'] == 'TMPO']
```

```python
# list of UniProt col values
col_uniprot = list(summary_df.UniProt)
# value corresponfing to TMPO gene
double_entries_list = col_uniprot[15].split(';')
# first UniProt entry
first_entry = double_entries_list[0]
# Second UniProt entry
second_entry = double_entries_list[1]
# updating UniProt column with just the first entry for TMPO gene 
col_uniprot[15] = first_entry
summary_df.UniProt = col_uniprot

# listing the TMPO row
row_to_update = list((summary_df.loc[15]))
# updating the UniProt value with second entry
row_to_update[4] = second_entry
# adding the new row to df
summary_df.loc[len(summary_df)] = row_to_update
```

```python
# now TMPO gene has 2 rows
summary_df[summary_df['Gene'] == 'TMPO']
```

# HGNC


```python
# HGNC dataset 
hgnc_df = pd.read_csv('hgnc-symbol-check.csv', header= 1)
```

```python
hgnc_df.head()
```

```python
# save approved genes
approved_genes = []
for index,row in hgnc_df.iterrows():
    if row['Match type'] == 'Approved symbol':
        approved_genes.append(row['Input'])
        
#check if all my input genes are approved
if len(set(approved_genes)) == len(set(summary_df.Gene)):
    print('All genes are approved! They are', len(set(approved_genes)))
else:
    print('Not all genes are approved. Check!')
```

```python
#read UniProt dataset
uni_df = pd.read_excel('uniprot-yourlist_M20191221E5A08BB0B2D1C45B0C7BC3B55FD2655669E875N+AND+organism--.xlsx')
uni_df.head()
```

```python
# dict: for each gene, uniprot entry
gene_uniprot = {}
for index,row in summary_df.iterrows():
    key = row['Gene']
    if key not in list(gene_uniprot.keys()):
        gene_uniprot[key] = [row['UniProt']]
    else:
        gene_uniprot[key].append(row['UniProt'])
        
# list of uniprot values
uniprot_list = [item for sublist in list(gene_uniprot.values()) for item in sublist]
```

```python
# filtering uniprot df by uniprot entries which appear also in summary df
final_df=uni_df.loc[uni_df['Entry'].isin(uniprot_list)]
final_df.head()
```

```python
#check lenght: ok because one gene has 2 different entries from summary df
len(final_df)
```

```python
#drop Gene names col: we have primal gene names, it's sufficient
final_df = final_df.drop(['Gene names'], axis=1)
```

```python
#renaming cols
original_cols = list(final_df.columns)
new_cols = ['Uniprot AC', 'Protein_Name', 'Organism', 'Gene_Symbol', 'Entrez_ID', 'Function']
final_df = final_df.rename(columns=dict(zip(original_cols, new_cols)))
# resetting indeces
final_df.reset_index(drop=True, inplace= True)
final_df.head()
```

```python
# check lenght without duplicates: ok
len(set(final_df.Gene_Symbol))
```

```python
# saving to csv
try:
    os.remove('data.csv')
except:
    pass

final_df.to_csv('data.csv')
```

# IID

```python
iid_df = pd.read_csv('PPIs.txt', sep='\t')
iid_df.head()
```

In this case we must use experimental data: if we find the tags: e.g. "experimental;predicted;ortolog", they ARE considered as experimental.

```python
# changing exp;ortho in exp
iid_df['evidence type'] = iid_df['evidence type'].replace('exp;ortho', 'exp')
#changing exp;ortho;pred in exp
iid_df['evidence type'] = iid_df['evidence type'].replace('exp;ortho;pred', 'exp')
#changing exp;pred in exp
iid_df['evidence type'] = iid_df['evidence type'].replace('exp;pred', 'exp')
```

### Results

We must consider also interactions among no-seed proteins.
Then I must consider the no-seed genes among the symbol 1 and symbol 2 cols and find their interactions.

```python
# list of genes in the first symbols col
symbols_1 = list(iid_df.symbol1)
# list of genes in the second symbols col
symbols_2 = list(iid_df.symbol2)

#verify that interactions involve always at least one seed gene
count = 0
for i in range(len(symbols_1)):
    if symbols_1[i] not in approved_genes and symbols_2[i] not in approved_genes:
        count+=1
    
print('We have in total ',len(iid_df), ' interactions. Among them, ', count,' are among no-seed proteins')
```

```python
tot_proteins = symbols_1 + symbols_2 
tot_proteins = list(set(tot_proteins))

seeds_iid_found = []
for gene in tot_proteins:
    if gene in approved_genes:
        seeds_iid_found.append(gene)
    
seeds_iid_found = list(set(seeds_iid_found))

print('We found ', len(seeds_iid_found), ' seed genes')
print('We have a total of ', len(tot_proteins), ' proteins')
```

# BIOGRID

Here we download the file of human organism from biogrid, and consider just the rows where appear at least one of our seed-genes. So on this way we should consider interactions among only seed-genes and among seed and no-seed genes.

Then we will check if there also interactions among only these no-seed genes.

```python
biogrid_df = pd.read_table('BIOGRID-ORGANISM-Homo_sapiens-3.5.179.tab2.txt', low_memory = False)
biogrid_df.head()
```

```python
seeds_interactions_A = biogrid_df[biogrid_df['Official Symbol Interactor A'].isin(approved_genes)]
seeds_interactions_B= biogrid_df[biogrid_df['Official Symbol Interactor B'].isin(approved_genes)]
seeds_interactions = pd.concat([seeds_interactions_A, seeds_interactions_B], axis = 0)
```

### Results

```python
A = list(set(list(seeds_interactions['Official Symbol Interactor A'])))
B = list(set(list(seeds_interactions['Official Symbol Interactor B'])))
tot_A_B = set(A + B)
tot_len = len(tot_A_B)

print('Total number of proteins (including seed genes): ', tot_len)
```

```python
no_seed_genes = []
for i, row in seeds_interactions.iterrows():
    if row['Official Symbol Interactor A'] not in approved_genes:
        no_seed_genes.append(row['Official Symbol Interactor A'])
    elif row['Official Symbol Interactor B'] not in approved_genes:
        no_seed_genes.append(row['Official Symbol Interactor B'])
no_seed_genes = list(set(no_seed_genes))
print('Number of No-Seed Genes: ', len(set(no_seed_genes))) 
n_seeds = tot_len - len(no_seed_genes)
print('Number of Seed Genes: ', n_seeds)
```

```python
c = 0
symbols_A = list(seeds_interactions['Official Symbol Interactor A'])
symbols_B = list(seeds_interactions['Official Symbol Interactor B'])
for i in range(len(seeds_interactions)):
    if symbols_A[i] in no_seed_genes and symbols_B[i] in no_seed_genes:
        c +=1
print("We have in total ", len(seeds_interactions), 'interactions. Among them, ', c,' are among no-seed genes')
```

```python
# save to csv
biogrid_interactions = biogrid_df[['Official Symbol Interactor A', 'Official Symbol Interactor B']]
try:
    os.remove('biogrid_interactions.csv')
except:
    pass

biogrid_interactions.to_csv('biogrid_interactions.csv')
```
